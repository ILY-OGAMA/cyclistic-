```python
# merge the data sets into one data frame
data = [Nov_2020, Dec_2020, Jan_2021, Feb_2021, Mar_2021, Apr_2021, May_2021, Jun_2021, Jul_2021, Aug_2021, Sep_2021, Oct_2021]
```

```python
cyclist_data = pd.concat(data)
```

```python
# add a new columns date, month, and year this will allow us to aggregare data for each day, month or year
cyclist_data['date'] = pd.to_datetime(cyclist_data['started_at'])
cyclist_data['year'] = pd.DatetimeIndex(cyclist_data['started_at']).year
cyclist_data['month'] = pd.DatetimeIndex(cyclist_data['started_at']).month
cyclist_data['day_of_week'] = pd.DatetimeIndex(cyclist_data['started_at']).weekday
```

```python
# inspect the new data frame
cyclist_data.info()
```

```python
# change the data types for started at and ended at to datetime
cyclist_data['started_at'] = cyclist_data['started_at'].astype('datetime64')
cyclist_data['ended_at'] = cyclist_data['ended_at'].astype('datetime64')
```

```python
# inspect the data frame to confirm the changes 
cyclist_data.info()
```

```python
# create a new calculated field ride time to get the ride lenght for each trip
cyclist_data['ride_time'] = cyclist_data['ended_at'] - cyclist_data['started_at'].shift(1)
```

```python
# inspect the new data frame
cyclist_data.head()
```

```python
# checking for missing values
cyclist_data.isnull().sum()
```

```python
# sort the data frame by missing values 
cyclist_data.isnull().sort_values(by ="start_station_name",ascending = False)
```
